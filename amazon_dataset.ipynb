```python
# Import libaries 
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# import the amazon CSV file

df = pd.read_csv('amazon.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['discounted_price'] = df['discounted_price'].str.replace("₹",'').str.replace(",",'').astype('float64')

df['actual_price'] = df['actual_price'].str.replace("₹",'').str.replace(",",'').astype('float64')
```

```python
# Convert the 'discounted_price' and 'actual_price' columns to USD. The exchange rate is 0.012.

exchange_rate = 0.012
df['discounted_price'] = df['discounted_price'].apply(lambda x: x * exchange_rate)
df['actual_price'] = df['actual_price'].apply(lambda x: x * exchange_rate)

# Use the round() command to round the prices to 2 decimal places.

df['discounted_price'] = round(df['discounted_price'], 2)
df['actual_price'] = round(df['actual_price'], 2)
```

```python
# Create a new column called discount_amount

df['discount_amount'] = df['actual_price'] - df['discounted_price']
```

```python
# Chnage the value ' | ' to 3.9 in the rating column
df.loc[df['rating'] == '|']

df['rating'] = df['rating'].str.replace('|', '2.0').astype('float64')
```

```python
# change the 'discount_percentage' column to a number 

df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64')
```

```python
# change numeric columns to numeric

numeric_col = ['discounted_price', 'actual_price', 'rating', 'rating_count']

for col in numeric_col:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df = df.drop_duplicates()
```

```python
df['product_id'].str.strip()
```

```python
# Split the 'category' column

splitcategory = df['category'].str.split('|', expand = True)

# rename the columns

splitcategory = splitcategory.rename(columns = {0:'category', 
                                                1:'subcategory'})
```

```python
replace_dict = {
    '&': ' & ',
    'MusicalInstruments':"Musical Instruments", 
    'OfficeProducts': 'Office Products',
    'HomeImprovement': 'Home Improvement',
    'NetworkingDevices': 'Networking Devices',
    'HomeTheater': 'Home Theater',
    'HomeAudio': 'Home Audio',
    'WearableTechnology': 'Wearable Technology',
    'ExternalDevices': 'External Devices',
    'DataStorage': 'Data Storage',
    'GeneralPurposeBatteries': 'General Purpose Batteries',
    'BatteryChargers': 'Battery Chargers',
    'OfficePaperProducts': 'Office Paper Products',
    'CraftMaterials': 'Craft Materials',
    'OfficeElectronics': 'Office Electronics',
    'PowerAccessories': 'Power Accessories',
    'HomeAppliances': 'Home Appliances',
    'AirQuality': 'Air Quality',
    'HomeStorage': 'Home Storage',
    'CarAccessories': 'Car Accessories',
    'ExternalDevices': 'External Devices',
    'DataStorage': 'Data Storage',
    'GeneralPurposeBatteries': 'General Purpose Batteries',
    'BatteryChargers': 'Battery Chargers',
    'OfficePaperProducts': 'Office Paper Products',
    'CraftMaterials': 'Craft Materials',
    'OfficeElectronics': 'Office Electronics',
    'PowerAccessories': 'Power Accessories',
    'HomeAppliances': 'Home Appliances',
    'AirQuality': 'Air Quality',
    'HomeStorage': 'Home Storage',
    'CarAccessories': 'Car Accessories',
    'HomeMedicalSupplies': 'Home Medical Supplies'
}
```

```python
# Replace the values in the 'category' column

splitcategory['category'] = splitcategory['category'].replace(replace_dict)

# Merge the splitcategory dataframe with the original dataframe

df = pd.concat([df, splitcategory], axis = 1)
```

```python
df.head()
```

```python
df = df.drop(columns = 'category')

# Add the new category column and subcategory column into the df dataframe.
df['category'] = splitcategory['category']
df['subcategory'] = splitcategory['subcategory']
```

```python
df["rating"].unique()
```

```python
ranking = []

for score in df['rating']:
    if score <= 2.0 : ranking.append('Poor')
    elif score <= 4.0 : ranking.append('Average')
    elif score <= 4.5 : ranking.append('Good')
    else : ranking.append('Excellent')

# Add the ranking list as a ranking column in the df dataframe.
# Use the .astype() command to convert the values in the ranking column into a category data type.

df['ranking'] = ranking

df['ranking'] = df['ranking'].astype('category')
```
